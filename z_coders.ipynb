```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("application_metadata.csv")
data.head()
```

```python
# Data Inspection & Cleaning
print(data.info())
print("\nSummary statistics:\n", data.describe())
print("\nMissing values:\n", data.isnull().sum())
print("\nNumber of duplicate rows:", data.duplicated().sum())
```

```python
data = pd.read_parquet("credit_history.parquet")
data.head()
```

```python
# Data Inspection & Cleaning
print(data.info())
print("\nSummary statistics:\n", data.describe())
print("\nMissing values:\n", data.isnull().sum())
print("\nNumber of duplicate rows:", data.duplicated().sum())
```

```python
data['num_delinquencies_2yrs'] = data['num_delinquencies_2yrs'].fillna(0)
```

```python
# Kredit score tarqatilishi
plt.hist(data['credit_score'], bins=40)
```
